```python
import requests
from bs4 import BeautifulSoup as soup
import csv
```

```python
headers = {'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
 'Accept-Encoding': 'gzip, deflate, br',
 'Accept-Language': 'en-GB,en;q=0.9,en-US;q=0.8,ml;q=0.7',
 'Cache-Control': 'max-age=0',
 'Connection': 'keep-alive',
 'Host': 'www.yellow-pages.ph',
 'Upgrade-Insecure-Requests': '1',
 'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.140 Safari/537.36'
 }
```

```python
#html = requests.get('https://www.yellow-pages.ph/search/government-offices/metro-manila/page-1',headers=headers)
html = requests.get('https://www.yellow-pages.ph/search/government-offices/metro-manila/page-2',headers=headers)
bsobj=soup(html.content,'lxml')
bsobj
```

```python
names=bsobj.find('div',{'class':'search-container-center'})
names
name=names.find('a',{'class':'yp-click'})
name.text.strip()
```

```python
governments_office = []
for names in bsobj.findAll('div',{'class':'search-business-name'}):
    for name in names.findAll('a',{'class':'yp-click'}):
        governments_office.append(name.text.strip())

governments_office
```

```python
description= []
for o_description in bsobj.findAll('span',{'class':'search-capsule-rounded search-badge'}):
        description.append(o_description.getText())

description
```

```python
phone_number = []
location=[]
for numbers in bsobj.findAll('div',{'class':'d-flex'}):
    for num in numbers.findAll('a',{'class':'btn btn-yp-default mr-2 biz-btn-call yp-click'}):
        phone_number.append(num['href'].strip())

for num in phone_number:
    print (num)
```

```python
adress = []
for o_adress in bsobj.findAll('span',{'class':'ellipsis'}):
        adress.append(o_adress.getText())

adress
```

```python
print("Writing scraped data to C:\\Users\\OMAR SAMB\\Desktop\\webscrapping\\bureaux_gouvernementaux.csv")
with open('C:\\Users\\OMAR SAMB\\Desktop\\webscrapping\\bureaux_gouvernementaux.csv', 'a',newline="") as csvfile:
    fieldnames = ['Bureau_Gouvernementaux','Description', 'Téléphone', 'Localisation']
    writer= csv.writer(csvfile,delimiter=';')
    #writer.writerow(fieldnames)
    for a in range(len(governments_office)):
        writer.writerow([governments_office[a],description[a],phone_number[a],adress[a]])
```
